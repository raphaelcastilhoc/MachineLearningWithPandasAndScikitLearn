```python
def plot_corr(df, size=25):
    corr = df.apply(lambda x: x.factorize()[0]).corr()
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
```

```python
plot_corr(df)
```

```python
from sklearn.model_selection import train_test_split
feature_columns = ["age", "workclass", "fnlwgt", "education-num", 
                        "marital-status", "occupation", "relationship", "race",
                        "sex", "capital-gain", "capital-loss", "hours-per-week",
                        "native-country"]
class_column = ["income"]
```

```python
x = (df['age'].map(str) + " ¬ " + df['workclass'].map(str) + " ¬ " + df['fnlwgt'].map(str)
         + " ¬ " + df['education-num'].map(str) + " ¬ " + df['marital-status'].map(str) + " ¬ " + df['occupation'].map(str)
         + " ¬ " + df['relationship'].map(str) + " ¬ " + df['race'].map(str) + " ¬ " + df['sex'].map(str)
         + " ¬ " + df['capital-gain'].map(str) + " ¬ " + df['capital-loss'].map(str) + " ¬ " 
         + df['hours-per-week'].map(str) + " ¬ " + df['native-country'].map(str)).values

y = df[class_column].values

split_test_percentage = 0.30

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = split_test_percentage, random_state=42)
```

```python
print("{0:0.2f}% in traning set".format(len(x_train)/len(df.index) * 100))
print("{0:0.2f}% in test set".format(len(x_test)/len(df.index) * 100))
```

```python
print("Original True  : {0} ({1:0.2f}%)".format(len(df.loc[df['income'] == 1]), (len(df.loc[df['income'] == 1])/len(df.index)) * 100.0))
print("Original False : {0} ({1:0.2f}%)".format(len(df.loc[df['income'] == 0]), (len(df.loc[df['income'] == 0])/len(df.index)) * 100.0))
print("")

# Check if "make more than 50k true and false" were splited correctly in 70% and 30% on Training data
print("Training True  : {0} ({1:0.2f}%)".format(len(y_train[y_train[:] == 1]), (len(y_train[y_train[:] == 1])/len(y_train) * 100.0)))
print("Training False : {0} ({1:0.2f}%)".format(len(y_train[y_train[:] == 0]), (len(y_train[y_train[:] == 0])/len(y_train) * 100.0)))
print("")

# Check if "make more than 50k true and false" were splited correctly in 70% and 30% on Test data
print("Test True      : {0} ({1:0.2f}%)".format(len(y_test[y_test[:] == 1]), (len(y_test[y_test[:] == 1])/len(y_test) * 100.0)))
print("Test False     : {0} ({1:0.2f}%)".format(len(y_test[y_test[:] == 0]), (len(y_test[y_test[:] == 0])/len(y_test) * 100.0)))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

model = Pipeline([('vect', CountVectorizer()),
                  ('tfidf', TfidfTransformer()),
                  ('clf', MultinomialNB())
                 ])
model
```

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(model, x_train, y_train.ravel(), cv=10, scoring=['accuracy','f1_macro','f1_weighted'], return_train_score=True)
scores
```

```python
scores_df = pd.DataFrame(scores)
scores_df.mean() * 100
```

```python
scores_df.describe()
```

```python
model.fit(x_train, y_train.ravel())
```
